# Cleaning the questionnaire dataset

```python
#import package that would be used
import pandas as pd
import numpy as np
```

```python
#dataset that would be cleaned: questionnaire
qn = pd.read_csv('questionnaire80.csv')
```

```python
#Drop columns that does not work
qn_drop = qn.drop(columns=['HSAQUEX', 'DBQ229', 'OHQ033', 'PAAQUEX', 'SMAQUEX2', 'SMAQUEX.x', 'SMAQUEX.y', 'OCD390G'])
qn_drop.head()
```

```python
#Dummy variables to change
my_list1 = ['SLQ050', 'ALQ101', 'BPQ020', 'BPQ056', 'BPQ059','BPQ080','HSQ500','HSQ510','HSQ520','HSQ571','HSQ590','DIQ160','DIQ170','DIQ172','DIQ180','DIQ050','CBQ596','CBQ505','CBQ535','CBQ550','CBQ552','CBQ580','DLQ010','DLQ020','DLQ040','DLQ050','DLQ060','DLQ080','FSD151','HEQ010','HEQ030','HIQ011','HUQ071','HUQ090','INQ020','INQ012','INQ030','INQ060','INQ080','INQ090','INQ132','INQ140','INQ150','MCQ010','MCQ053','MCQ070','MCQ080','MCQ082','MCQ086','MCQ092','MCQ160A','MCQ160N','MCQ160B','MCQ160C','MCQ160D','MCQ160E','MCQ160F','MCQ160G','MCQ160G','MCQ160M','MCQ160K','MCQ160L','MCQ160O','MCQ203','MCQ220','MCQ300A','MCQ300B','MCQ300C','MCQ365A','MCQ365B','MCQ365C','MCQ365D','MCQ370A','MCQ370B','MCQ370C','MCQ370D','OHQ770','PFQ049','PFQ051','PFQ054','PFQ057','PFQ090','PAQ605','PAQ620','PAQ635','PAQ650','PAQ665','PUQ100','PUQ110','SLQ060','SMQ020','SMQ856','SMQ866','SMQ870','SMQ872','SMQ874','SMQ878','SMQ681','SMQ851','SMQ863','SMDANY', 'WHQ070']

my_list2 = ['DPQ010', 'DPQ020', 'DPQ030', 'DPQ040', 'DPQ060', 'DPQ070', 'DPQ080', 'DPQ090']
```

```python
#New clean list
qn_drop
```

```python
#New name
qn_new = qn_drop
```

```python
# only change 5 to 0 
linear_type2 = ['HSD010', 'DBQ700', 'HUQ010', 'OHQ835']

qn_new['HSD010'] = np.where((qn_new.HSD010 == 5), 0, qn_new.HSD010)
```

```python
qn_new['HSD010']
```

```python
#Linear type 1 
	##Change 777777, 999999 and missing into NaN, nothing else

linear_type1 = ['CBD070', 'CBD090', 'CBD110', 'CBD120', 'CBD130']

for i in linear_type1:
    qn_new.replace({i:777777}, np.nan, inplace=True) 

for i in linear_type1:
    qn_new.replace({i:999999}, np.nan, inplace=True) 



#Linear type 2
	##Change 5 into 0, change 4 into 1, change 3 into 2, change 2 into 3, change 1 into 4, 9 and missing into NaN

linear_type2 = ['HSD010', 'DBQ700', 'HUQ010', 'OHQ835']


for i in linear_type2:
    qn_new.replace({i:9}, np.nan, inplace=True) 



#Linear type 3
	##1 into 2, 2 into 0, 3 into 1, 7, 9 and missing into NaN

linear_type3 = ['DIQ010', 'WHO030', 'WHO040', 'HUQ020', 'CBQ545', 'CBQ590', 'FSDHH', 'FSDAD', 'FSQ165', 'INFMMPC']



for i in linear_type3:
    qn_new.replace({i:7}, np.nan, inplace=True) 

for i in linear_type3:
    qn_new.replace({i:9}, np.nan, inplace=True) 




#Linear type 4
	##4, 7, and missing into NaN

linear_type4 = ['DBQ197']

for i in linear_type4:
    qn_new.replace({i:4}, np.nan, inplace=True) 

for i in linear_type4:
    qn_new.replace({i:7}, np.nan, inplace=True) 

for i in linear_type4:
    qn_new.replace({i:9}, np.nan, inplace=True) 


#Linear type 5
	##change 5555 into 22, change 7777, 9999 and missing into NaN

linear_type5 = ['DBD895']

for i in linear_type5:
    qn_new.replace({i:5555}, 22, inplace=True) 

for i in linear_type5:
    qn_new.replace({i:7777}, np.nan, inplace=True) 

for i in linear_type5:
    qn_new.replace({i:9999}, np.nan, inplace=True) 



#Linear type 6
	##change 7777, 9999 and missing into NaN

linear_type6 = ['DBD905', 'DBD910', 'PAD680', 'WHD010', 'WHD020', 'WHQ050', 'WHD140', 'WHD050']

for i in linear_type6:
    qn_new.replace({i:7777}, np.nan, inplace=True) 

for i in linear_type6:
    qn_new.replace({i:9999}, np.nan, inplace=True) 




#Linear type 10
	##2 into 0, 3 into 2

linear_type10 = ['HUQ030']

for i in linear_type10:
    qn_new.replace({i:2}, 0, inplace=True) 

for i in linear_type10:
    qn_new.replace({i:3}, 2, inplace=True) 




#New Binary type 1
	##2 into 1, 3 into 0, 7, 9 and missing into NaN

binary_type1 = ['IMQ011', 'IMQ020']

for i in binary_type1:
    qn_new.replace({i:2}, 1, inplace=True) 

for i in binary_type1:
    qn_new.replace({i:3}, 0, inplace=True) 

for i in binary_type1:
    qn_new.replace({i:7}, np.nan, inplace=True) 

for i in binary_type1:
    qn_new.replace({i:9}, np.nan, inplace=True) 



#Linear type 12
	##77, 99 and missing into NaN

linear_type12 = ['IND235', 'SLD010H', 'HUQ051', 'OHQ030', 'HUQ041']

for i in linear_type12:
    qn_new.replace({i:77}, np.nan, inplace=True) 

for i in linear_type12:
    qn_new.replace({i:99}, np.nan, inplace=True) 






#Linear type 16
	##change 8 into 0, 77, 99 and missing into NaN

linear_type16 = ['PAQ710', 'PAQ715']

for i in linear_type16:
    qn_new.replace({i:8}, 0, inplace=True) 

for i in linear_type16:
    qn_new.replace({i:77}, np.nan, inplace=True) 

for i in linear_type16:
    qn_new.replace({i:99}, np.nan, inplace=True) 



#Linear type 17
	##777, 999, and missing into NaN

linear_type17 = ['SMD460', 'HOD050']

for i in linear_type17:
    qn_new.replace({i:777}, np.nan, inplace=True) 

for i in linear_type17:
    qn_new.replace({i:999}, np.nan, inplace=True) 



#Linear type 18
	##77777, 99999 and missing into NaN

linear_type18 = ['WHQ150']

for i in linear_type18:
    qn_new.replace({i:77777}, np.nan, inplace=True) 

for i in linear_type18:
    qn_new.replace({i:99999}, np.nan, inplace=True) 
```

```python
qn_linear = qn_new
```
